```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np

# ----------------------------
# Custom Dataset
# ----------------------------
class JokeDataset(Dataset):
    def __init__(self, df, vocab_size):
        self.user_ids = df['user_id'].values.astype(np.int64)
        self.jokes = df[[f'second_bestjoke', 'third_bestjoke', 'fourth_bestjoke', 'fifth_bestjoke']].values
        self.labels = df['bestjoke'].values
        self.vocab_size = vocab_size

    def __len__(self):
        return len(self.user_ids)

    def __getitem__(self, idx):
        user_id = self.user_ids[idx]

        # Flatten joke vectors
        joke_feats = np.concatenate([self.jokes[idx][i] for i in range(4)])
        label = self.labels[idx]
        return torch.tensor(user_id), torch.tensor(joke_feats, dtype=torch.float32), torch.tensor(label, dtype=torch.float32)
```

```python
import torch
import torch.nn as nn

# ----------------------------
# Neural Network
# ----------------------------
class JokeNet(nn.Module):
    def __init__(self, num_users, vocab_size, emb_dim=10):
        super().__init__()
        self.user_emb = nn.Embedding(num_users, emb_dim)
        self.joke_encoder = nn.Linear(vocab_size, emb_dim)

        self.mlp = nn.Sequential(
            nn.Linear(5 * emb_dim, 50),
            nn.ReLU(),
            nn.Linear(50, 30),
            nn.ReLU(),
            nn.Linear(30, vocab_size)
        )
        self._init_weights()

    def _init_weights(self):
        for layer in self.mlp:
            if isinstance(layer, nn.Linear):
                nn.init.kaiming_normal_(layer.weight, nonlinearity='relu')

    def forward(self, user_id, joke_features):
        user_vec = self.user_emb(user_id)  # (B, 10)
        joke_vecs = [self.joke_encoder(joke_features[:, i * self.joke_encoder.in_features: (i + 1) * self.joke_encoder.in_features])
                     for i in range(4)]  # List of 4 (B, 10)
        joke_vecs = torch.cat(joke_vecs, dim=1)  # (B, 40)
        all_feats = torch.cat([user_vec, joke_vecs], dim=1)  # (B, 50)
        return self.mlp(all_feats)
```

```python
import torch
import torch.optim as optim
import torch.nn as nn

# ----------------------------
# Train Function
# ----------------------------
def train_model(model, dataloader, epochs=10, lr=1e-3):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    optimizer = optim.Adam(model.parameters(), lr=lr)
    criterion = nn.MSELoss()

    for epoch in range(epochs):
        model.train()
        total_loss = 0

        for user_ids, joke_feats, labels in dataloader:
            user_ids = user_ids.to(device)
            joke_feats = joke_feats.to(device)
            labels = labels.to(device)

            optimizer.zero_grad()
            outputs = model(user_ids, joke_feats)

            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()

        avg_loss = total_loss / len(dataloader)
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {avg_loss:.4f}")

    return model
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

# ---- Model Definition ----
class JokeMLP(nn.Module):
    def __init__(self, vocab_size, embedding_dim=16, hidden_dims=[64, 32]):
        super(JokeMLP, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)

        self.fc1 = nn.Linear(embedding_dim * 4, hidden_dims[0])
        self.fc2 = nn.Linear(hidden_dims[0], hidden_dims[1])
        self.output = nn.Linear(hidden_dims[1], vocab_size)

        self.init_weights()

    def init_weights(self):
        nn.init.kaiming_normal_(self.fc1.weight, nonlinearity='relu')
        nn.init.kaiming_normal_(self.fc2.weight, nonlinearity='relu')
        nn.init.kaiming_normal_(self.output.weight, nonlinearity='linear')

        for bias in [self.fc1.bias, self.fc2.bias, self.output.bias]:
            nn.init.zeros_(bias)

    def forward(self, x):
        emb = self.embedding(x)  # (batch_size, 4, embedding_dim)
        flat = emb.view(x.size(0), -1)  # (batch_size, 4 * embedding_dim)
        x = F.relu(self.fc1(flat))
        x = F.relu(self.fc2(x))
        return self.output(x)  # logits

# ---- Example Dataset ----
class JokeDataset(Dataset):
    def __init__(self, df):
        self.inputs = df[['secondbestjoke', 'thirdbestjoke', 'fourthbestjoke', 'fifthbestjoke']].values
        self.targets = df['bestjoke'].values

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        x = torch.tensor(self.inputs[idx], dtype=torch.long)
        y = torch.tensor(self.targets[idx], dtype=torch.long)
        return x, y

# ---- Training Function ----
def train_model(model, dataloader, epochs=20, lr=1e-3):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.CrossEntropyLoss()

    for epoch in range(epochs):
        model.train()
        total_loss = 0
        for x, y in dataloader:
            x, y = x.to(device), y.to(device)

            optimizer.zero_grad()
            logits = model(x)
            loss = criterion(logits, y)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
        print(f"Epoch {epoch+1}/{epochs} | Loss: {total_loss/len(dataloader):.4f}")

    return model
```

```python
import pandas as pd

# Assuming you already have top5_per_user dataframe
# Sort by user_id and rating to ensure top jokes are on top for each user
top5_per_user_sorted = top5_per_user.sort_values(by=['user_id', 'rating'], ascending=[True, False])

# Generate the best jokes for each user
top5_per_user_sorted['rank'] = top5_per_user_sorted.groupby('user_id').cumcount() + 1

# Pivot to get jokes as columns per user
top5_per_user_wide = top5_per_user_sorted.pivot(index='user_id', columns='rank', values='joke_id')

# Rename columns to match the joke ranks
top5_per_user_wide.columns = ['bestjoke', 'secondbestjoke', 'thirdbestjoke', 'fourthbestjoke', 'fifthbestjoke']

# Reset index to get user_id as a regular column
top5_per_user_wide.reset_index(inplace=True)

# Now, top5_per_user_wide contains the desired format
top5_per_user_wide
```

```python
joke_cols = ['secondbestjoke', 'thirdbestjoke', 'fourthbestjoke', 'fifthbestjoke']

# One-hot encode each column individually
encoded_parts = [pd.get_dummies(top5_per_user_wide[col], prefix=col) for col in joke_cols]

# Combine all one-hot encodings horizontally
one_hot_encoded = pd.concat(encoded_parts, axis=1)

# If you want to keep user_id along with it:
final_df = pd.concat([top5_per_user_wide[['user_id']],top5_per_user_wide[['bestjoke']], one_hot_encoded], axis=1)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader

# Extract input features and target
X = final_df.drop(columns=['user_id', 'bestjoke']).astype(float)
y = final_df['bestjoke']

# Encode target labels (bestjoke)
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X.values, y_encoded, test_size=0.2, random_state=42
)

# Convert to tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)

train_loader = DataLoader(TensorDataset(X_train_tensor, y_train_tensor), batch_size=64, shuffle=True)
```

```python
class JokeEmbeddingNN(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(JokeEmbeddingNN, self).__init__()
        self.hidden = nn.Linear(input_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.output = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        x = self.relu(self.hidden(x))
        return self.output(x)

    def get_embedding(self, x):
        return self.relu(self.hidden(x))

# Hyperparams
input_dim = X.shape[1]
hidden_dim = 64  # embedding size
output_dim = len(np.unique(y_encoded))  # number of unique bestjokes

model = JokeEmbeddingNN(input_dim, hidden_dim, output_dim)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
epochs = 10
for epoch in range(epochs):
    model.train()
    for xb, yb in train_loader:
        optimizer.zero_grad()
        preds = model(xb)
        loss = criterion(preds, yb)
        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
```

```python
model.eval()
with torch.no_grad():
    all_embeddings = model.get_embedding(torch.tensor(X.values, dtype=torch.float32)).numpy()

# Create a DataFrame for embeddings
embedding_df = pd.DataFrame(all_embeddings)
embedding_df['user_id'] = final_df['user_id'].values
```
